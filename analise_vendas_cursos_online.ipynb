#### 2. Resolução do Desafio

- Calcular a receita total gerada pela venda dos cursos.
- Identificar o curso com o maior número de vendas.
- Visualizar a distribuição das vendas ao longo do tempo através de gráficos.

```python
#--------- Código para visualização da distribuição de vendas por curso e identificação do curso mais vendido----------

# Agrupando os nomes do curso e somando as quantidades de vendas
df_vendas_curso = df.groupby(by=['Nome do Curso'])['Quantidade de Vendas'].sum().reset_index()

# Ordenando o dataframe resultante do mais vendido para o menos vendido
df_vendas_curso_ordenado = df_vendas_curso.sort_values('Quantidade de Vendas', ascending=False)

# Criando as variáveis x e y para plotagem do gráfico
x = df_vendas_curso_ordenado['Nome do Curso']
y = df_vendas_curso_ordenado['Quantidade de Vendas']

# Plotagem do gráfico de barras usando o matplotlib
plt.bar(x, y)
plt.xlabel('Nome do Curso')
plt.ylabel('Quantidade de Vendas')
plt.title('Quantidade de Vendas por Curso')
plt.xticks(rotation=90)
plt.show()

# Resetando o índice do dataframe ordenado
df_vendas_curso_ordenado.reset_index(inplace=True)

# Obtendo o nome do curso mais vendido pela posição do íncide
curso_mais_vendido = df_vendas_curso_ordenado['Nome do Curso'][0]

print(f'O curso mais vendido foi o de {curso_mais_vendido}')
```

```python
#--------- Código para visualização da dispersão entre a quantidade de vendas e o preço unitário do curso ----------
sns.scatterplot(data=df, x = 'Quantidade de Vendas', y = 'Preço Unitário')
```

```python
#--------- Código para medir a correlação entre as variáveis de quantidade de vendas e o preço unitário do curso ----------
correlacao_vendas_preco = df['Quantidade de Vendas'].corr(df['Preço Unitário'])
print(f'A correlação entre a quantidade de vendas e o preço unitário é de {correlacao_vendas_preco:.2f}')
```

```python
#--------- Código para visualizar a distribuição da quantidade de vendas ao longo do tempo ----------

# Agrupando as datas de venda e somando as quantidades de vendas
df_venda_periodo = df.groupby(by=['Data'])['Quantidade de Vendas'].sum().reset_index()

# Ordenando a coluna de data em ordem crescente
df_venda_periodo_ordenado = df_venda_periodo.sort_values('Data', ascending=True)

# Criando as variáveis x e y para plotagem do gráfico
x = df_venda_periodo_ordenado['Data']
y = df_venda_periodo_ordenado['Quantidade de Vendas']

# Plotagem do gráfico de barras usando o matplotlib
plt.figure(figsize=(12,4))
plt.plot(x, y)
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.title('Quantidade de Vendas por Data')
plt.show()
```

```python
# Criando a coluna de faturamento através da multiplicação entre quantidade de vendas e preço unitário
df['Faturamento'] = df['Quantidade de Vendas'] * df['Preço Unitário']
```

```python
#--------- Código para visualizar a distribuição faturamento ao longo do tempo ----------

# Agrupando as datas de venda e somando os valores de faturamento
df_faturamento_periodo = df.groupby(by=['Data'])['Faturamento'].sum().reset_index()

# Ordenando a coluna de data em ordem crescente
df_faturamento_periodo_ordenado = df_faturamento_periodo.sort_values('Data', ascending=True)

# Criando as variáveis x e y para plotagem do gráfico
x = df_faturamento_periodo_ordenado['Data']
y = df_faturamento_periodo_ordenado['Faturamento']

# Plotagem do gráfico de barras usando o matplotlib
plt.figure(figsize=(12,4))
plt.plot(x, y)
plt.xlabel('Data')
plt.ylabel('Faturamento')
plt.title('Faturamento por Data')
plt.show()
```

```python
# identificando a soma do faturamento total do dataset
faturamento_total = df['Faturamento'].sum()
print(f'O faturamento total foi de {faturamento_total:.2f}')
```
